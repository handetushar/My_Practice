```python
final_df=pd.merge(df, df_country, on='Country Code', how='left')
```

```python
country_names=final_df.Country.value_counts().index
```

```python
plt.pie(values[:3],labels=country_names[:3], autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Conclusion
1. Rating is between 1.8-2.4 --->Poor
2. Rating is between 2.5-3.4 --->Average
3. Rating is between 3.5-3.9 --->Good
4. Rating is between 4.0-4.4 --->Very good
5. Rating is between 4.5-4.9 --->Excellent


```python
plt.rcParams['figure.figsize']=(12, 6)
sns.barplot(x='Aggregate rating', y="Rating count", hue='Rating color', data=ratings, palette=['white', 'red','orange', 'yellow', 'green'])
```

## Conclusion
1. Not rated is very high
2. Maximum rating betwen 2.5 to 3.4


```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green'])
```
